## Question 1: Car Matrix Generation 

```python
def generate_car_matrix(dataframe):
    id_1 = dataframe['id_1'].unique()
    id_2 = dataframe['id_2'].unique()
    id_1.sort()
    id_2.sort()
    new_dataframe = pd.DataFrame(data=None,index=id_1,columns=id_2)
    for i in range(len(dataframe)):
        id1 = dataframe.iloc[i].id_1
        id2 = dataframe.iloc[i].id_2
        car = dataframe.iloc[i].car
        new_dataframe.at[id1,id2] = car
    for i in range(len(new_dataframe)):
        new_dataframe.iat[i,i]=0.0
    return(new_dataframe)
    
```

```python
generate_car_matrix(dataframe)
```

## Question 2: Car Type Count Calculation 

```python
def get_type_count(dataframe):
    dataframe['car_type'] = 'Unknown'
    dataframe.loc[dataframe['car'] <=15,'car_type'] = 'low'
    dataframe.loc[(dataframe['car']>15) & (dataframe['car']<=25),'car_type'] = 'medium'
    dataframe.loc[dataframe['car']>25,'car_type'] = 'high'
    car_types = dataframe['car_type'].value_counts()
    car_types_dictionary = dict(car_types)
    keys = list(car_types_dictionary.keys())
    keys.sort()
    sorted_car_types = {i:car_types_dictionary[i] for i in keys}
    return(sorted_car_types)
```

```python
get_type_count(dataframe)
```

## Question 3: Bus Count Index Retrieval 

```python
def get_bus_indexes(dataframe):
    mean_bus_value = dataframe['bus'].mean()
    twice_mean_bus_value = 2*mean_bus_value
    bus_value_indexes = dataframe.index[dataframe['bus'] > twice_mean_bus_value].tolist()
    return(bus_value_indexes)
```

```python
get_bus_indexes(dataframe)
```

## Question 4: Route Filtering 

```python
def filter_routes(dataframe):
    average_truck_value = dataframe['truck'].mean()
    filtered_routes = dataframe[dataframe['truck'] > average_truck_value]
    routes = filtered_routes['route'].tolist()
    routes.sort()
    return(routes)
    
```

```python
filter_routes(dataframe)
```

## Question 5: Matrix Value Modification

```python
def multiply_matrix(matrix_dataframe):
    for i in range(len(matrix_dataframe)):
        
        for j in range(len(matrix_dataframe)):
            value_frame = matrix_dataframe.iat[i,j]
            if(value_frame > 20):
                value_frame*=0.75
            else:
                value_frame*=1.25
            value_frame_rounded = round(value_frame,1)
            matrix_dataframe.iat[i,j] = value_frame_rounded
    return(matrix_dataframe)
```

```python
car_matrix = generate_car_matrix(dataframe)
multiply_matrix(car_matrix)
```

## Question 6: Time Check 

```python
def time_check(dataframe):
    weekday_number = {'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'Friday':4,'Saturday':5,'Sunday':6}
    dataframe['startDay_number'] = dataframe['startDay'].map(weekday_number)
    dataframe['endDay_number'] = dataframe['endDay'].map(weekday_number)
    dataframe['valid_timestamp'] = (dataframe['startTime'] < dataframe['endTime']) & (dataframe['startDay_number'] < dataframe['endDay_number'])
    arrays = [np.array(dataframe['id']),np.array(dataframe['id_2'])]
    s = pd.Series(data=dataframe['valid_timestamp'],index=arrays)
    for i in range(len(s)):
        s.iloc[i] = True
    return(s)
```
