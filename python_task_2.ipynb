## Question 1: Distance Matrix Calculation 

```python
def calculate_distance_matrix(dataframe):
    id_start = dataframe['id_start'].unique()
    distance_dataframe = pd.DataFrame(data=None,index=id_start,columns=id_start)
    distance_series = dataframe['distance'].copy()
    cumsum_distance = distance_series.cumsum()
    dataframe['cumdistance'] = cumsum_distance
    for i in range(len(dataframe)):
        start_index = dataframe.iloc[i].id_start
        end_index = dataframe.iloc[i].id_end
        distance_value = dataframe.iloc[i].distance
        distance_dataframe.at[start_index,end_index] = distance_value
        distance_dataframe.at[end_index,start_index] = distance_value
    for i in range(len(dataframe)):
        id_start = dataframe.iloc[i].id_start
        for j in range(i+1,len(dataframe)):
            id_end = dataframe_3.iloc[j].id_end
            cumulative_distance = dataframe_3.iloc[j].cumdistance
            distance_dataframe.at[id_start,id_end] = cumulative_distance 
            distance_dataframe.at[id_end,id_start] = cumulative_distance
    for i in range(len(distance_dataframe)):
        distance_dataframe.iat[i,i] = 0.0
    return(distance_dataframe)
        
```

## Question 2: Unroll Distance Matrix 

```python
def unroll_distance_matrix(dataframe):
    distance_columns = dataframe.columns
    distance_index = dataframe.index
    data = []
    for i in range(len(dataframe)):
        index = distance_index[i]
        for j in range(len(dataframe)):
            col = distance_columns[j]
            value_distance = dataframe.iloc[i].iloc[j]
            if(index != col):
                data.append([index,col,value_distance])
    unrolled_dataframe = pd.DataFrame(data=data,columns=['start_id','end_id','distance'])
    return(unrolled_dataframe)
```

## Question 3: Finding IDs within Percentage Threshold 

```python
def find_ids_within_ten_percentage_threshold(dataframe,reference_value):
    distance_reference = 0
    distances = []
    for i in range(len(dataframe)):
        startid = dataframe.iloc[i].start_id
        if(startid == reference_value):
            distances.append(dataframe.iloc[i].distance)
    mean_distance_reference = sum(distances)/len(distances)
    ten_percent_threshold1 = mean_distance_reference*0.1 + mean_distance_reference
    ten_percent_threshold2 = mean_distance_reference - mean_distance_reference*0.1
    ids=[]
    for i in range(len(dataframe)):
        startid = dataframe.iloc[i].start_id
        distance_value = dataframe.iloc[i].distance
        end_id = dataframe.iloc[i].end_id
        if(distance_value <= ten_percent_threshold1 and distance_value >= ten_percent_threshold2 and startid!=reference_value):
            ids.append(startid)
    ids.sort()
    unique_ids = np.unique(ids)
    return(unique_ids)
```

```python
import datetime

def calculate_time_based_toll_rates(dataframe):
    records_list = []
    for i in range(len(dataframe)):
        record = dataframe.iloc[i]
        for j in range(4):
            records_list.append(record)
    time_series_dataframe = pd.DataFrame(data=records_list,columns=dataframe.columns)
    day_list = [['Monday','Friday'],['Tuesday','Saturday'],['Wednesday','Sunday'],['Saturday','Sunday']]
    day_list = day_list*4
    time = datetime.datetime(2020,6,17,23,59,59)
    time2 = datetime.datetime(2020,6,17,0,0,0)
    time3 = datetime.datetime(2020,6,17,18,0,0)
    time4 = datetime.datetime(2020,6,17,10,0,0)
    midnight = time.strftime("%X")
    next_day = time2.strftime("%X")
    evening = time3.strftime("%X")
    morning = time4.strftime("%X")
    time_dataframe = [['Monday','Friday',next_day,morning],['Tuesday','Saturday',morning,evening],['Wednesday','Sunday',evening,midnight],['Saturday','Sunday',next_day,midnight]]
    time_dataframe = time_dataframe*1806
    time = pd.DataFrame(data=time_dataframe,columns=['start_day','end_day','start_time','end_time'])
    time_series_dataframe.reset_index(inplace=True)
    time_series_dataframe.drop('index',axis=1,inplace=True)
    time_series_dataframe['start_day'] = time['start_day'].copy()
    time_series_dataframe['end_day'] = time['end_day'].copy()
    time_series_dataframe['start_time'] = time['start_time'].copy()
    time_series_dataframe['end_time'] = time['end_time'].copy()
    modified_values = []
    for i in range(len(time_series_dataframe)):
        startday = time_series_dataframe.iloc[i].start_day
        enday = time_series_dataframe.iloc[i].end_day
        startime = time_series_dataframe.iloc[i].start_time
        endtime = time_series_dataframe.iloc[i].end_time
        moto_value = time_series_dataframe.iloc[i].moto
        car_value = time_series_dataframe.iloc[i].car
        rv_value = time_series_dataframe.iloc[i].rv
        bus_value = time_series_dataframe.iloc[i].bus
        truck_value = time_series_dataframe.iloc[i].truck
        if(startday == 'Monday' or startday =='Tuesday' or startday =='Wednesday'):
        
            if(startime == '00:00:00' and endtime =='10:00:00'):
                moto_value*=0.8
                car_value*=0.8
                rv_value*=0.8
                bus_value*=0.8
                truck_value*=0.8
            elif(startime == '10:00:00' and endtime =='18:00:00'):
                moto_value*=1.2
                car_value*=1.2
                rv_value*=1.2
                bus_value*=1.2
                truck_value*=1.2
            else:
                moto_value*=0.8
                car_value*=0.8
                rv_value*=0.8
                bus_value*=0.8
                truck_value*=0.8
        else:
                moto_value*=0.7
                car_value*=0.7
                rv_value*=0.7
                bus_value*=0.7
                truck_value*=0.7
        modified_moto_value = round(moto_value,2)
        modified_car_value = round(car_value,2)
        modified_rv_value = round(rv_value,2)
        modified_bus_value = round(bus_value,2)
        modified_truck_value = round(truck_value,2)
        modified_values.append([modified_moto_value,modified_car_value,modified_rv_value,modified_bus_value,modified_truck_value])
    modified_values_dataframe = pd.DataFrame(data=modified_values,columns=['moto','car','rv','bus','truck'])
    time_series_dataframe.drop(columns=['moto','car','rv','bus','truck'],inplace=True)
    time_series_dataframe['moto'] = modified_values_dataframe['moto'].copy()
    time_series_dataframe['car'] = modified_values_dataframe['car'].copy()
    time_series_dataframe['rv'] = modified_values_dataframe['rv'].copy()
    time_series_dataframe['bus'] = modified_values_dataframe['bus'].copy()
    time_series_dataframe['truck'] = modified_values_dataframe['truck'].copy()
    return(time_series_dataframe)
```
